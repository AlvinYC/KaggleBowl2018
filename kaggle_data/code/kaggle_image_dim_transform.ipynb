```python
# webstie https://nucleisegmentationbenchmark.weebly.com
# paper   https://drive.google.com/file/d/0ByERBiBsEbuTOEJISEpwSkR0SlE/view
import os
import re
import PIL
import time
import random
import shutil
import itertools
import collections
import numpy as np
import re

import matplotlib.pyplot as plt
from PIL import Image
from pathlib import Path
```

```python

DATASET_DIR = Path('/home/alvin/cei/notebook_home/kaggleBowl2018/unet_label/Unet_fix_ShannChing_v2/')


#DATASET_DIR = Path('../../../kaggle_data/stage1_train/')
DATASET_DIR_LIST = sorted(list(DATASET_DIR.glob('*')))
region_info = []
mask_num = 0
for data_dir in DATASET_DIR_LIST:
    if not os.path.isdir(str(data_dir)): continue
    
    img_name = str(data_dir.stem) + '.png'
    
    
    ################################################################
    #                  kaggle_dir/images
    ################################################################
    img_dir = data_dir / 'images' / img_name
    im = np.array(Image.open(img_dir))
    
    title = '==========================================================\n'
    title += img_name + str(im.shape)
    print(title)
    
    if len(im.shape) == 2:
        print('this png is GRAY, please make sure GIMP image->mode is RGB')
        continue
    if len(im.shape) == 3 and im.shape[2] not in [3,4]:
        print('this png channel number have problem, please check image shape')
        continue
    if len(im.shape) == 3 and im.shape[2] == 4:
        # [m,n,4] case, need remove alpha
        shape_orig = im.shape
        im = im[:,:,:3]
        shape_new = im.shape
        im_new = Image.fromarray(im)
        im_new.save(img_dir,"PNG")
        print(str(shape_orig) + ' --> ' + str(shape_new) + 'remove alpha completed')
        
    ################################################################
    #                  kaggle_dir/masks
    ################################################################    
    mask_dir = data_dir / 'masks'
    mask_dir_list = sorted(list(mask_dir.glob('*.png')))   
    
    
    mask_dir_string = [str(this_string) for this_string in mask_dir_list]
    mask_num += len(mask_dir_string)
    
    for mask_path_a in mask_dir_string:
        short_path_a = re.sub('.*/|\.png','',mask_path_a)

        mask_img = np.array(Image.open(mask_path_a))
        #print('\t'+ short_path_a + '\t'+str(mask_img.shape))
        if len(mask_img.shape) != 2:
            # not gary format [m,n] case, select channel 0 as new image
            shape_orig = mask_img.shape
            mask_img = mask_img[:,:,0]
            shape_new = mask_img.shape
            im_new = Image.fromarray(mask_img,'L')
            im_new.save(mask_path_a,"PNG")
            print(short_path_a + '\t' + str(shape_orig) + '-->' + str(shape_new) + ' to gray')
        else:
            shape_orig = mask_img.shape
            print(short_path_a + '\t' + str(shape_orig))
```
